# House Prices Predict by Tensor-flow

```python
train_df = pd.read_csv('train.csv') #(1460, 81)
test_df =  pd.read_csv('test.csv')
all_df = pd.concat((train_df.iloc[:, 1:-1], test_df.iloc[:, 1:])) #2919 rows × 79 columns (remove SalePrice)
#all_features #2919, 79


print(train_df.shape)
print(test_df.shape)
print(all_df.shape)
all_df.head()
```

# Data Preprocessing

```python
#standardizing
numeric_features = all_df.dtypes[train_df.dtypes != 'object'].index
# After standardizing the data all means vanish, hence we can set missing
# values to 0
all_df[numeric_features] = all_df[numeric_features].fillna(0)

#encode others
all_features = pd.get_dummies(all_df, dummy_na=True) #(2919, 331)


all_features
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0, 1))
# Scale both the training inputs
scaled_data_x = scaler.fit_transform(all_features)
scaled_train_y = scaler.fit_transform(train_df[['SalePrice']])
```

```python
scaled_data_x_df = pd.DataFrame(scaled_data_x, columns=all_features.columns.values)
scaled_data_x_df
```

# build our tf.keras model

```python
model = tf.keras.Sequential()

#
model.add(tf.keras.layers.Dense(50, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(50, activation='relu'))
model.add(tf.keras.layers.Dense(1))

#
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
X_train_and_test = scaled_data_x_df.values

#
X_train = X_train_and_test[:1460]
Y_train = scaled_train_y

X_test = X_train_and_test[1460:]
```

```python
# Train the model
model.fit(
    X_train,
    Y_train,
    epochs=50,
    shuffle=True,
    verbose=0
)
```

```python

test_error_rate = model.evaluate(X_train, Y_train, verbose=0)
print("The mean squared error (MSE) for the test data set is: {}".format(test_error_rate))
```

```python
prediction = model.predict(X_test)
print(prediction.shape)
prediction
```

```python
def rev_min_max_func(scaled_val):
    max_val = max(train_df['SalePrice'])
    min_val = min(train_df['SalePrice'])
    og_val = (scaled_val*(max_val - min_val)) + min_val
    return og_val
```

```python
Y_test = rev_min_max_func(prediction)
print(Y_test.shape)
Y_test
```

```python
test_df['SalePrice'] = Y_test
test_df
```
